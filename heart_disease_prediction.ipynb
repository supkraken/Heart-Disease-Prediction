```python
import numpy as np  #for array/lists type operations
import pandas as pd #for data frame : tabular structured table operations
from sklearn.model_selection import train_test_split #split original data into train and test data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # used to evaluate model : performance
```

DATA COLLECTION AND PROCESSING

```python
# IMPORTING DATASET AS DATAFRAME
heart_data = pd.read_csv("heart_disease_data.csv")
```

```python
# printing first 5 rows of dataset
heart_data.head()
```

```python
# printing last 5 rows of dataset
heart_data.tail()
```

```python
# number of columns and rows
heart_data.shape
```

```python
# getting info about dataset
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
#  no null values in dataset
```

```python
# statistical measures abt data
heart_data.describe()
```

```python
# 0 - no defect
#  1 - heart defect
#  in target field
# checking distribution of target var
heart_data['target'].value_counts()
```

```python
# splitting features and target
x = heart_data.drop(columns='target',axis=1)
y = heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

we need to feed the x and y to the machine learning algorithms

we need to divide the data into training data and test data

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2,)
#  x_train stores features of training data
#  y_train stores target of training data
#  x_test stores features of test data
#  y_test stores target of test data
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
print(y.shape , y_train.shape , y_test.shape)
```

MODEL TRAINING

```python
# logistic regression
model = LogisticRegression()
```

```python
# training logisticregression model with training data
model.fit(x_train,y_train)
```

```python
# evaluating our model how well it is performing
# using accuracy score
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy on training data : ',training_data_accuracy)
# giving 85% accuracy score
```

```python
# accuracy on training data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy on test data : ',testing_data_accuracy)
# giving 81% accuracy score
```

```python
# BUILDING A PREDICTIVE SYSTEM
input_data = (67,1,0,160,286,0,0,108,1,1.5,1,3,2)
# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)
# reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print("The person has no Heart Disease")
else :
  print("The person has a Heart Disease")
```
